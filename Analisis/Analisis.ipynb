```python
fundas = provedores[provedores['Objeto contractual'].str.contains("CADAVER")]
fundas.head()
```

```python
fundas.describe()
```

Segun fuentes de informacion, las fundas de cadaver rondean un valor de $12 en promedio

```python
mediaFundas = 12
def categoria_sobreprecio(x):
    sobre = x['Precio unitario']/mediaFundas *100
    return sobre

fundas['% Sobreprecio']=fundas.apply(categoria_sobreprecio,axis=1)
fundas.head()
```

#### Segmentación por empresas

```python
acc3 = acc_1.groupby(["Nombre"]).sum()
acc3 = acc3.reset_index()

m = acc3['Capital Invertido'].mean()

acc6 = acc_1.groupby(["Empresa",'Nombre'])['Capital Invertido'].sum()

acc6 = pd.DataFrame(acc6)
acc6 = acc6.loc[acc6['Capital Invertido']<=m]
acc6 = acc6.reset_index()
acc6.head()
```

```python
acc6['%'] = acc6.groupby('Empresa')['Capital Invertido'].transform(lambda x: (x/sum(x))*100 )
acc6 = acc6.round(3)
```

### En esta sección se analiza la relación entre los accionistas de una empresa y la cantidad de veces que esta es contratada como proveedora

```python
accionistas_actuales = accionistas_actuales[accionistas_actuales['Expediente'].notna()]
accionistas_actuales['count_nombre'] = accionistas_actuales['Nombre'].map(accionistas_actuales['Nombre'].value_counts())
accionistas_actuales_join = accionistas_actuales[["Nombre", "count_nombre"]]
```

```python
accionistas_actuales_join.drop_duplicates(
                     keep = "first", inplace = True)
```

```python
lista_proveedores = pd.read_csv('contrataciones_por_proveedor.csv',delimiter=';')

lista_proveedores['veces_contratada'] = lista_proveedores['Proveedor'].map(lista_proveedores['Proveedor'].value_counts())
```

```python
personas_accionistas_actuales = lista_proveedores.merge(
    accionistas_actuales_join, left_on='Proveedor', right_on='Nombre',
    how="right", 
)
```

```python
personas_accionistas_actuales = personas_accionistas_actuales.rename(columns={'Nombre': 'Proveedor', 'count_nombre': 'empresas_accionista_actual'})
```

```python
personas_accionistas_actuales[["veces_contratada", "empresas_accionista_actual",]].corr(
    method="spearman"
)
```

```python
personas_accionistas_actuales[["veces_contratada", "Precio total",]].corr(
    method="spearman"
)
```

```python
sns.distplot(personas_accionistas_actuales["veces_contratada"], color="m")
plt.title('Distribucion de veces que un proveedor ha sido contratado')
```

```python
sns.set(style="darkgrid")

tips = sns.load_dataset("tips")
g = sns.jointplot("veces_contratada", "Precio total", data=personas_accionistas_actuales,
                  kind="reg", truncate=False,
                  color="m", height=7)

plt.title('Veces que un proveedor ha sido contratado vs monto de contrato')
```
